```python
from keras.preprocessing.text import Tokenizer
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Embedding, LSTM
```

```python
from sklearn.model_selection import train_test_split
```

```python
from keras.utils.np_utils import to_categorical
```

```python
import re
```

```python
#ML
```

```python
data = pd.read_csv(r'Sentiment.csv')

# Keeping only the neccessary columns
data = data[['text','sentiment']]
print(len(data))
```

```python
data
```

```python

data['text'] = data['text'].apply(lambda x: x.lower())
data['text'] = data['text'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))

#print(data[ data['sentiment'] == 'Positive'].size)
#print(data[ data['sentiment'] == 'Negative'].size)

for idx,row in data.iterrows():
    row[0] = row[0].replace('rt',' ')
    
batch_size = 5000
tokenizer = Tokenizer(nb_words=batch_size, split=' ')

tokenizer.fit_on_texts(data['text'].values)
X = tokenizer.texts_to_sequences(data['text'].values)

maxx = 0
c=0
for ll in X:
    c=len(ll)
    if c>maxx:
        maxx=c
print('Max',maxx)
X = pad_sequences(X)
print(len(X[0]))
```

```python
data
```

```python
embed_dim = 128
lstm_out = 128
model = Sequential()
model.add(Embedding(batch_size, embed_dim,input_length = X.shape[1], dropout=0.2))
model.add(LSTM(lstm_out,return_sequences=True ,dropout_U=0.2, dropout_W=0.2))
model.add(LSTM(32,return_sequences=True))
model.add(LSTM(32))
model.add(Dense(3,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam',metrics = ['accuracy'])
print(model.summary())
```

```python
print(data['sentiment'].values)
Y = pd.get_dummies(data['sentiment']).values
print(Y)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.33, random_state = 42)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
model.fit(X_train, Y_train, nb_epoch = 30, batch_size=32)
```
